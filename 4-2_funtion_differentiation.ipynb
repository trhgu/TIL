```python
# 그래프를 통해 기울기(slope) 살표보기


def f(x):
    return x**3 - 3 * x**2 + x


x = np.linspace(-1, 3, 400)  # linspace(A,b,c) A 에서 B 까지 C 토막 낸다 결과물은 float 타입
y = f(x)

plt.plot(x, y)

plt.plot(-0.5, f(-0.5), 'go')
plt.plot(x, (3*(-0.5)**2-6*(-0.5)+1)*x+1, "g--")

plt.plot(0, 0, 'ro')
plt.plot(x, x, 'r-.')

plt.plot(1, -1, 'bo')
plt.plot(x, (3*1**2-6*1+1)*(x-1)-1, "b-")

plt.plot(1.5, f(1.5), 'co')
plt.plot(x, (3*(1.5)**2-6*(1.5)+1)*x, "c:")

plt.plot(2, f(2), 'mo')
plt.plot(x, (3*(2)**2-6*(2)+1)*x-4, "m-")


plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3, 6))
plt.yticks(np.arange(-4, 2))


plt.annotate('', xy=(1, 0), xytext=(0, 0), arrowprops=dict(facecolor='gray'))
plt.annotate('예시', xy=(1, 1), xytext=(1, 0), arrowprops=dict(facecolor='gray'))


plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("함수의 기울기")
plt.show()
```

```python
# derivative() 수치미분
from scipy.misc import derivative

x = np.linspace(-1, 3, 400)  # linspace(A,b,c) A 에서 B 까지 C 토막 낸다 결과물은 float 타입
y = f(x)


plt.plot(x, y)

plt.plot(-0.5, f(-0.5), 'go')
plt.plot(x, derivative(f, 0.5, dx=1e-6)*x+1, "g--")
plt.plot(0, 0, 'ro')
plt.plot(x, derivative(f, 0, dx=1e-6)*x, 'r-.')

plt.plot(1, -1, 'bo')
plt.plot(x, derivative(f, 1, dx=1e-6)*(x-1)-1, "b-")

plt.plot(1.5, f(1.5), 'co')
plt.plot(x, derivative(f, 1.5, dx=1e-6)*x, "c:")

plt.plot(2, f(2), 'mo')
plt.plot(x, derivative(f, 2, dx=1e-6)*x-4, "m-")


plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3, 6))
plt.yticks(np.arange(-4, 2))

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("함수의 기울기")
plt.show()
```

### 심파이

심파이는 심볼릭 연산(symbolic operation)을 지원하는 파이썬 패키지다. 심볼릭 연산이란 사림아 연필로 계사하는 미분/적분과 동일한 형태의 연산을 말한다. 즉 $x^2$의 미분 연산을 수행하면 그 결과가 $2x$란 형태로 출력된다. 딥 러닝(deep learning)등에 많이 사용되는 파이썬의 tensorflow패키지나 pytorch 패키지도 기울기 함수 계산을 위해 이러한 심볼릭 연산 기능을 갖추고 있다.

```python
import sympy

# Juypter 노트북에서 수학식의 LaTeX 표현을 위해 필요함
sympy.init_printing(use_latex='mathjax')
```

심볼릭 연상에서 사용하는 심볼릭 변수(symbolic variable)는 일반 프로그래밍에서 사용하는 변수와 다르다. 일반 프로그래밍에서 사용하는 변수는 이미 메모리에 씌여 있는 어떤 숫자를 기호로 쓴 것에 지나지 않지만 심볼릭 변수는 아무런 숫자도 대입이 되어 있지 않다. 따라서 $x^2$의 미분 연산을 수행하려면 우선 SymPy의 `symbols()` 명령을 사용하여 $x$라는 기호가 단순한 숫자나 벡터 변수가 아닌 심볼(symbols)임을 알려주어야 한다. 이렇게 정의된 심볼 변수는 `Symbil` 클래스 자료형이 된다.

```python
# 심볼릭 변수로 변환
x = sympy.symbols('x')

# 복수도 가능
x, y = sympy.symbols('x y')
x, y
```
